## KMeans sur l'embedding des phrases de BERT

```python
from sklearn.cluster import KMeans
kmeansmodel=KMeans(n_clusters=2,n_init=100,max_iter=500)
start=time.time()
kmeans=kmeansmodel.fit(df_BERT)
end=time.time()
print('Durée :',round((end-start)/60,2),' minutes')
```

```python
df_BERT['kmeans_pred']=kmeans.predict(df_BERT)
```

## Gaussian Mixture sur l'embedding BERT

```python
from sklearn.mixture import GaussianMixture
```

```python
bic = []
lowest_bic = np.infty
n_components_range = range(1, 7)
cv_types = ['spherical', 'tied', 'diag', 'full']
for cv_type in cv_types:
    for n_components in n_components_range:
        # Fit a Gaussian mixture with EM
        gmm = GaussianMixture(n_components=n_components,
                                      covariance_type=cv_type)
        gmm.fit(df_BERT.iloc[:,:-2])
        bic.append(gmm.bic(df_BERT.iloc[:,:-2]))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm
```
